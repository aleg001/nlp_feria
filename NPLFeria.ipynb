```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
texts, labels = zip(*data)
labels = [label_mapping[label] for label in labels]

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.3, random_state=42)

# Vectorizamos los textos usando TF-IDF
vectorizer = TfidfVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Entrenamos el modelo supervisado (Logistic Regression)
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vect, y_train)
```

```python
# Función para realizar predicciones con el modelo combinado (BART + modelo supervisado)
def combined_prediction(text):
    # Predicción del modelo supervisado (con TF-IDF)
    vect_text = vectorizer.transform([text])
    supervised_pred = model.predict(vect_text)  # Devuelve una lista de predicciones

    # Predicción del modelo Zero-Shot con BART
    bart_labels = ["left-wing", "right-wing", "libertarian", "authoritarian"]
    bart_result = classifier(text, bart_labels)
    bart_pred = bart_labels[bart_result['labels'].index(max(bart_result['labels'], key=lambda x: bart_result['scores'][bart_result['labels'].index(x)]))]

    # Devuelve ambas predicciones
    return supervised_pred[0], bart_pred

```

```python
y_pred_combined = [combined_prediction(text) for text in X_test]
```

```python
accuracy = accuracy_score(y_test, y_pred_combined)
print("\nReporte de clasificación del modelo combinado:")
print(classification_report(y_test, y_pred_combined))
```

```python
inverse_label_mapping = {v: k for k, v in label_mapping.items()}
```

```python
def preprocess_text(text):
    text = text.lower()
    text = text.replace("estado", "gobierno")
    text = text.replace("colectividad", "comunidad")
    return text
```

```python
# Función para análisis avanzado usando NLP
def advanced_nlp_analysis(response):
    response_doc = nlp(response)

    # Extraemos las entidades relevantes (gobierno, empresa, etc.)
    entities = [ent.text for ent in response_doc.ents if ent.label_ in ["ORG", "GPE", "PERSON", "MONEY", "NORP"]]

    # Analizamos el sentimiento general de la respuesta
    sentiment_result = sentiment_analyzer(response)[0]
    sentiment_score = sentiment_result['score'] if sentiment_result['label'] == "POSITIVE" else -sentiment_result['score']

    # Detectamos temas importantes a partir de palabras clave y frases (gobierno, economía, propiedad)
    keywords = [token.lemma_ for token in response_doc if token.pos_ in ["NOUN", "ADJ", "VERB"]]

    return {
        "entities": entities,
        "sentiment": sentiment_score,
        "keywords": keywords
    }
```

```python
# Función de reglas personalizadas que aprovecha el NLP
def custom_rules_advanced(response, ideology_scores):
    # Preprocesamos el texto y aplicamos el análisis NLP avanzado
    response = preprocess_text(response)
    nlp_results = advanced_nlp_analysis(response)

    # Analizamos las entidades encontradas (si se menciona "gobierno", "empresa", etc.)
    if "gobierno" in nlp_results['entities']:
        ideology_scores["left-wing"] += 0.15
        ideology_scores["authoritarian"] += 0.10
    if "empresa" in nlp_results['entities']:
        ideology_scores["right-wing"] += 0.10

    # Ajustamos puntajes basados en el sentimiento: más positivo hacia un lado, más negativo hacia otro
    if nlp_results['sentiment'] > 0:
        ideology_scores["left-wing"] += 0.10
    else:
        ideology_scores["right-wing"] += 0.10

    # Detectamos palabras clave y temas: por ejemplo, "redistribución" aumenta el puntaje hacia la izquierda
    if "redistribuir" in nlp_results['keywords'] or "igualdad" in nlp_results['keywords']:
        ideology_scores["left-wing"] += 0.10
    if "libertad" in nlp_results['keywords']:
        ideology_scores["libertarian"] += 0.10
    if "propiedad" in nlp_results['keywords']:
        ideology_scores["right-wing"] += 0.10

    # Ajustamos más reglas basadas en keywords importantes
    if "colectividad" in nlp_results['keywords'] or "comunidad" in nlp_results['keywords']:
        ideology_scores["authoritarian"] += 0.10

    return ideology_scores
```

```python
# Función para hacer preguntas y obtener respuestas del usuario
def ask_questions():
    responses = []
    print("Por favor responde las siguientes preguntas:")
    for question in questions:
        response = input(question + "\n ")
        responses.append(response)
    return responses
```

```python
def analyze_responses(responses):
    ideology_scores = {label: 0 for label in labels}

    for response in responses:
        response = preprocess_text(response)

        # Usamos el modelo combinado (supervisado + BART)
        supervised_pred, bart_pred = combined_prediction(response)

        # Mapeamos las predicciones a las etiquetas correspondientes
        bart_label_idx = label_mapping[bart_pred]

        # Aumentamos los puntajes según las predicciones de ambos modelos
        ideology_scores[labels[supervised_pred]] += 0.5  # Damos igual peso al modelo supervisado
        ideology_scores[labels[bart_label_idx]] += 0.5  # Damos igual peso a BART Zero-Shot

        # Aplicamos reglas personalizadas avanzadas con NLP para ajustar los puntajes
        ideology_scores = custom_rules_advanced(response, ideology_scores)

    # Normalizamos los puntajes
    for label in ideology_scores:
        ideology_scores[label] /= len(questions)

    return ideology_scores
```

```python
# Función para mostrar los resultados con el nombre de la ideología en lugar de valores numéricos
def display_ideology_scores(ideology_scores):
    print("\nTus puntajes ideológicos son:")
    for label, score in ideology_scores.items():
        # Mapeamos el índice de la etiqueta de vuelta al nombre de la ideología
        ideology_name = inverse_label_mapping[labels.index(label)]
        print(f"{ideology_name}: {score:.2f}")
```

```python
# Función para graficar los resultados de la ideología
def plot_ideology(ideology_scores):
    plt.figure(figsize=(10, 6))
    plt.barh(list(ideology_scores.keys()), list(ideology_scores.values()), color='skyblue')
    plt.xlabel('Puntaje')
    plt.title('Tu Posición Ideológica')
    plt.xlim(0, 1)
    for i, (label, score) in enumerate(ideology_scores.items()):
        plt.text(score, i, f"{score:.2f}", va='center')
    plt.show()
```

```python
# Función principal para ejecutar el flujo
def main():
    responses = ask_questions()
    ideology_scores = analyze_responses(responses)

    # Mostrar los resultados en formato de texto
    print("\nTus puntajes ideológicos son:")
    for label, score in ideology_scores.items():
        print(f"{label}: {score:.2f}")

    # Graficar los resultados
    plot_ideology(ideology_scores)
```

```python
# Ejecutar la función principal
main()
```
